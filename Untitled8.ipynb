```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Открыть файл 
df = pd.read_csv('/content/drive/MyDrive/test_task/Тестовое задание - tz_data.csv')
df.head()
```

```python
# Информация по данным таблицы
print(df.info())
print('-'*55)
print('Количество уникальных значений:\n')
print(df.nunique())
print('-'*55)
print('Пропущенные значения:\n')
print(df.isnull().sum())
print('-'*55)
print('Статистика для численных данных:\n')
print(df.describe())
```

```python
# Удалить строки с пустыми значениями
df.dropna(subset=None, inplace=True)
df.isnull().sum()
```

```python
# Удалить столбец "good (1)"
df = df.drop('good (1)', axis=1)
df.head()
```

```python
df.iloc[[98]]
```

```python
# Удалить строку, содержащую не верно значение, для построения диаграммы рассеяния
df = df.drop(index=[99])
```

```python
# Перевод столбца "у" в формат float
df['y'] = df['y'].astype(float)
```

# **Вариант 1**

```python
# Строим диаграммы рассеяния для каждой области
sns.relplot(x='x', 
            y='y', 
            col='area', 
            hue='cluster_name', 
            col_wrap=3, 
            data=df);
```

# **Вариант 2**

```python
# Добавить колонку с цветами
clusters = df['cluster_name'].unique()                                          # названия кластеров
colors = ['blue', 'yellow', 'green', 'red']                                     # список цветов
color_dict = dict(zip(clusters, colors))                                        # словарь кластеров и цветов
df['color'] = df['cluster_name'].map(color_dict)                                # создаём столбец "color" и добавляем в него цвета
df.to_csv('/content/drive/MyDrive/test_task/new.csv', index=False)              # cохранение данных в CSV файле
df.head()
```

```python
# Создаём папку для сохранения диаграмм
import os
path = '/content/drive/MyDrive/test_task/scatters/'
os.mkdir(path)
```

```python
# Задаём цвета для легенды
blue = plt.scatter(x=df['x'], y=df['y'], color=colors[0])
yellow = plt.scatter(x=df['x'], y=df['y'], color=colors[1])
green = plt.scatter(x=df['x'], y=df['y'], color=colors[2])
red = plt.scatter(x=df['x'], y=df['y'], color=colors[3])
```

```python
 # Функция для построения диаграммы рассеяния
def png(name):

  df_area = df.loc[df['area'] == name]
  
  plt.figure(figsize=(15,15))
  plt.scatter(x=df_area['x'], y=df_area['y'], s=100, c=df_area['color'], edgecolors='black')
  plt.title(name ,fontsize=30)

  annotations = list(df_area['keyword'])
  X = list(df_area['x'])
  Y = list(df_area['y'])
  for i, label in enumerate(annotations):
    plt.text(X[i], Y[i]+0.2, label)

  plt.legend((blue, yellow, green, red),
            (clusters),
            scatterpoints=1, 
            title='Кластеры',
            loc='upper right',
            bbox_to_anchor=(1.1, 0.85), 
            ncol=1, 
            fontsize=8)

  plt.axis('off')
  plt.savefig(path+name)
  plt.show()
```

```python
# Создаём диаграмму для каждой области
areas = df['area'].unique()
for area in areas:
  png(area)
  print('-'*148)
```
